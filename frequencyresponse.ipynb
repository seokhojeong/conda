```python
import numpy as np
import matplotlib.pyplot as plt
```

## Basic parameters

```python
m = 1.
k = 10.
zeta = 0.05
wn = np.sqrt(k/m)
wd = wn*np.sqrt(1.-zeta**2.)
```

## Evaluate the normalised frequency response
\begin{equation*}
\frac{H_u(\omega)}{(u_{st})_o} = \frac{1}{[1-(\omega/\omega_n)^2]+i[2\zeta (\omega/\omega_n)]}
\end{equation*}

```python
def freqresponse(freqratio,dampingratio):
    "A frequency response function of a damped SDOF system"
    h = 1./((1.-freqratio**2.)+2.j*dampingratio*freqratio) # normalised frequency response
    return h

dt = 0.01 # define the time step
N = 8192 # total number of points
df = 1/(N*dt)
freq = df*np.arange(N) # define the frequency array
freqratio = 2.*np.pi*freq/wn # frequency ratio

H = freqresponse(freqratio, zeta)
```

## Plot the frequency response

```python
#plt.plot(freqratio, np.abs(H))
plt.plot(freqratio, np.abs(freqresponse(freqratio, 0.05)), label='zeta=5%')
plt.plot(freqratio, np.abs(freqresponse(freqratio, 0.25)), label='zeta=25%')
plt.plot(freqratio, np.abs(freqresponse(freqratio, 0.707)), label='zeta=71%')
plt.xlim((0, 3))
plt.ylim((0, 5))
plt.legend()
```

```python
plt.plot(freqratio, -np.angle(freqresponse(freqratio, 0.05)), label='zeta=5%')
plt.plot(freqratio, -np.angle(freqresponse(freqratio, 0.25)), label='zeta=25%')
plt.plot(freqratio, -np.angle(freqresponse(freqratio, 0.707)), label='zeta=71%')
plt.xlim((0, 3))
plt.legend()
```
