```python
import numpy as np
import matplotlib.pyplot as plt
```

## Initial condition

```python
m = 1.
k = 10.
zeta = 0.05
wn = np.sqrt(k/m)
wd = wn*np.sqrt(1.-zeta**2.)


v0 = 1./m
u0 = 0.
```

## Evaluate the impulse response

```python
time = 0.01*np.arange(4096) # define the time array
tau = 5. # delay in time
h = 1./(m*wd)*np.exp(-zeta*wn*(time-tau))*np.sin(wd*(time-tau)) # inpulse response (time>=tau)
h[np.where(time<tau)]=0 # zero response (time < tau)
```

## Plot the impulse response

```python
plt.plot(time, h)
```
